# Gen RL Pipeline

```python
from automarl.components.rl.rl_pipeline import RLPipelineComponent
from automarl.serialization.json_component_utils import gen_component_from

rl_pipeline : RLPipelineComponent = gen_component_from(rl_pipeline_config)
```

```python
rl_pipeline.pass_input({
    "base_directory" : PATH_TO_STORE_EXPERIMENTS,
                        "artifact_relative_directory" : experiment_name,
                        "create_new_directory" : True,
                        "do_full_setup_of_seed" : True}
                        )

experiment_path = rl_pipeline.get_artifact_directory()

print(f"Experiment path: {experiment_path}")
```

# Do the training

```python
from automarl.components.loggers.global_logger import activate_global_logger

activate_global_logger(rl_pipeline.get_artifact_directory())
```

```python
rl_pipeline.run()
```

## See training results

```python
AGGREGATE_NUMBER = 4
```

```python

from automarl.components.loggers.result_logger import ResultLogger
from automarl.components.rl.trainers.rl_trainer.rl_trainer_component import RLTrainerComponent

rl_trainer : RLTrainerComponent = rl_pipeline.get_rl_trainer()

results_logger : ResultLogger = rl_trainer.get_results_logger()
```

```python
#results_logger.plot_graph(x_axis='episode', y_axis=[('total_reward', name)], to_show=False)
results_logger.plot_confidence_interval(x_axis='episode', y_column='episode_reward',show_std=True, to_show=False, y_values_label=experiment_name, aggregate_number=AGGREGATE_NUMBER)
results_logger.plot_linear_regression(x_axis='episode', y_axis='episode_reward', to_show=False, y_label=experiment_name + '_linear', save_path="reward_per_episode.svg")
```

### Per agent 

```python

from automarl.components.rl.trainers.agent_trainer.agent_trainer_component import AgentTrainer


for agent_name, agent_trainer in rl_trainer.get_agents_trainers().items():

    agent_trainer : AgentTrainer = agent_trainer

    agent_results_logger : ResultLogger = agent_trainer.get_results_logger()

    #results_logger.plot_graph(x_axis='episode', y_axis=[('total_reward', name)], to_show=False)
    agent_results_logger.plot_confidence_interval(x_axis='episode', y_column='episode_reward',show_std=True, to_show=False, y_values_label=experiment_name, aggregate_number=AGGREGATE_NUMBER)
    agent_results_logger.plot_linear_regression(x_axis='episode', y_axis='episode_reward', to_show=True, y_label=experiment_name + '_linear', save_path="reward_per_episode.svg")



```
